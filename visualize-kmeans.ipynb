```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_distance(x0,y0,x1,y1):
    d = np.sqrt((x1-x0)**2 + (y1-y0)**2)
    return d
```

```python
x = [2,2,8,5,7,6,1,4]
y = [10,5,4,8,5,4,2,9]
center_x = [2,5,1]
center_y = [10,8,2]
count = 0
while True:
    count += 1
    
    
    # assign objects
    classified_x = [[],[],[]]
    classified_y = [[],[],[]]
    for i in range(len(x)):
        d = np.inf
        for j in range(len(center_x)):
            tmp = calc_distance(x[i], y[i], center_x[j], center_y[j])
            if d > tmp:
                d = tmp
                label = j
        classified_x[label].append(x[i])
        classified_y[label].append(y[i])
    
    
    # Update cluster centroids
    center_x_updated = [np.mean(cx) for cx in classified_x]
    center_y_updated = [np.mean(cy) for cy in classified_y]
    
    
    # visualize
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.scatter(classified_x[0],classified_y[0], s=100, c='blue', label='class1')
    ax.scatter(classified_x[1],classified_y[1], s=100, c='green', label='class2')
    ax.scatter(classified_x[2],classified_y[2], s=100, c='orange', label='class3')
    ax.scatter(center_x,center_y, c='red', marker='^', label='cluster centroid')
    
    ax.legend()
    ax.set_title('count:{}'.format(count))
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    
    fig.show()
    # fig.savefig('a10-{}.png'.format(count), dpi=300)

    
    # 
    if center_x == center_x_updated and center_y == center_y_updated:
        print('done')
        break
    else:
        center_x = center_x_updated.copy()
        center_y = center_y_updated.copy()
```
